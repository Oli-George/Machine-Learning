```python
import random
```

```python

def rps_game(player1_choice, player2_choice):
    choices = ['rock', 'paper', 'scissors']
    print(f"\nChoose from: {choices}\n")
    player2_choice = input().lower().strip()

    player1_choice = random.choice(choices)

    if player2_choice not in choices:
        return "\nInvalid choice! Please choose Rock, Paper, or Scissors."

    if player1_choice == player2_choice:
        return "\nIt's a tie! You both chose " + player1_choice.capitalize()

    if (player1_choice == 'rock' and player2_choice == 'scissors') or \
       (player1_choice == 'paper' and player2_choice == 'rock') or \
       (player1_choice == 'scissors' and player2_choice == 'paper'):
        return "\nComputer wins! It chose " + player1_choice.capitalize()
    return "\nYou win! Computer chose " + player1_choice.capitalize()

# Run the game
print(rps_game(None, None))
```
